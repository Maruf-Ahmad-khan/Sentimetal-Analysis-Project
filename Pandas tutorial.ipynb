```python
import requests

# Define the URL and headers
url = 'https://pokkweb.affise.com/v2/statistics/conversions?date_from=2024-05-01&date_to=2024-05-26&timezone=Asia/Kolkata&slice=offer_id&page=2&limit=500'  # Replace with the actual API endpoint
headers = {
    'API-Key': '23df424b0a53b0899f78685966243ee61'
}

# Make the GET request with headers
response = requests.get(url, headers=headers)

# Print status code and headers for debugging
print("Status Code:", response.status_code)
print("Headers:", response.headers)

# Check if the request was successful
if response.status_code == 200:
    try:
        # Attempt to parse the response as JSON
        data = response.json()
        print("JSON Data:", data)
    except requests.JSONDecodeError as e:
        # Handle JSON decode error
        print("JSON Decode Error:", e)
        print("Response content:", response.text)
else:
    # Handle unsuccessful request
    print(f"Error: {response.status_code}")
    print("Response content:", response.text)
```

```python
import requests
import pandas as pd
import io

# Define the URL
url = 'http://report.pokkt.com/reports/videoBackendReport?startDateTime=2015-05-17+00%3A00%3A00&endDateTime=2015-05-17+23%3A59%3A59&campaignFormType=Video'

# Define the headers with API key
headers = {
    'API-Key': '23df424b0a53b0899f78685966243ee61'
}

# Make the GET request
response = requests.get(url, headers=headers)

# Check if the request was successful
if response.status_code == 200:
    # Read the response content into a pandas DataFrame
    data = pd.read_csv(io.StringIO(response.text))
    
    # Save the DataFrame to a CSV file
    data.to_csv('output.csv', index=False)
    print("Data has been successfully saved to 'output.csv'")
else:
    # Handle the error
    print(f"Error: {response.status_code}")
    print("Response content:", response.text)
```
